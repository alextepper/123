```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 21, 100)
y = x**2
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set(title = "Plot of y = x^2", xlabel = 'x', ylabel = 'y')
fig.savefig("fig1")
```

```python
df = pd.read_csv('raw.githubusercontent.com_justmarkham_DAT8_master_data_chipotle.tsv', sep = '\t')
df['item_price'] = df['item_price'].apply(lambda price: float(price[1:]))
df.head()
```

```python
fig, ax = plt.subplots()
ax.scatter(df.item_price, df.quantity)
ax.set(title = "Relationship between Item and quantity ordered", xlabel = 'Item price', ylabel = 'quantity ordered')
fig.savefig("fig2")
```

```python
x = df.groupby('item_name').sum('quantity').sort_values('quantity', ascending=False).head(5)
x = x.drop(columns = ['order_id', 'item_price'])
plt.bar(x.index, x['quantity'])
plt.xticks(rotation=30, ha='right')
plt.title('Top 5 items...')
plt.xlabel("item name")
plt.ylabel('Total quantity')
# x.plot(kind='bar')
```
