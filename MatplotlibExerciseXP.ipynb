```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]  # Prices
y = [5, 4, 3, 2, 1]  # Demand
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set(title = 'Product Demand vs Price',xlabel='Price', ylabel='Demand')
fig.savefig("Fig1")
```

```python
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm
fig, ax = plt.subplots()
ax.scatter(x,y)
ax.set(title = 'Children’s Height vs Age',xlabel='Age (years)', ylabel='Height (cm)')
fig.savefig("Fig2")
```

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
fig, ax = plt.subplots(1,2)
ax[0].plot(x,y1)
ax[1].scatter(x,y2)
ax[0].set(title="y=sin(x)")
ax[1].set(title="y=cos(x)")
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(x,y1)
ax[1].scatter(x,y2)
ax[0].legend("y=sin(x)")
ax[1].legend("y=cos(x)")
```
