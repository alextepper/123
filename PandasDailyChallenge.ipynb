```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('starbucks.csv')
df.head(10)
```

```python
df.sort_values(' Sugars (g)', ascending=False).head(1)
```

```python
df['Beverage_category'].nunique()
```

```python
df['Beverage_category'].mode()
```

```python
df.groupby('Beverage_category').mean('Calories')
```

```python
df['Caffeine (mg)'] = pd.to_numeric(df['Caffeine (mg)'], errors = 'coerce')
df['Caffein to calories'] = df['Caffeine (mg)']/df['Calories']
df.head()
```

```python
df.groupby('Beverage_category').mean('Caffein to calories').sort_values('Caffein to calories', ascending=False)
```

```python
plt.barh(df['Beverage_category'],df['Caffein to calories'])
```
